### Task 4

Calculate the values of the matrix  
$$
z(x, y) = (x - 5)^2 - (y - 5)^2 \quad \text{for} \quad x, y = 1, 2, \ldots, 10
$$

Draw and compare the plots of values z:

a) **contour plots** – `contour`, `contour3`  
b) **surface plots** – `surf` (`surfc`, `surfl`), `mesh` (`meshc`, `meshz`), **waterfall**

````python
from mpl_toolkits.mplot3d import Axes3D

x = np.array([x for x in range(1, 11)])
y = np.array([x for x in range(1, 11)])
matrix = np.zeros((10, 10))
for i in range(10):
    for j in range(10):
        matrix[i, j] = (x[i] - 5) ** 2 - (y[j] - 5) ** 2
print(f"Matrix:\n{matrix}")

# a) Contour plots
X, Y_grid = np.meshgrid(x, y)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
cp = axs[0].contour(X, Y_grid, matrix, levels=15)
axs[0].set_title('Contour plot')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
fig.colorbar(cp, ax=axs[0])
axs[0].grid(True)

ax3d = fig.add_subplot(1, 2, 2, projection='3d')
ax3d.contour3D(X, Y_grid, matrix, levels=15, cmap='viridis')
ax3d.set_title('Contour3D plot')
ax3d.set_xlabel('x')
ax3d.set_ylabel('y')
ax3d.set_zlabel('z')

plt.tight_layout()
plt.show()

# b) Surface plots
fig = plt.figure(figsize=(14, 5))
ax1 = fig.add_subplot(1, 3, 1, projection='3d')
surf = ax1.plot_surface(X, Y_grid, matrix, cmap='viridis')
ax1.set_title('Surface plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
fig.colorbar(surf, ax=ax1, shrink=0.5)

ax2 = fig.add_subplot(1, 3, 2, projection='3d')
mesh = ax2.plot_wireframe(X, Y_grid, matrix, color='orange')
ax2.set_title('Mesh plot')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')

ax3 = fig.add_subplot(1, 3, 3, projection='3d')
ax3.plot_surface(X, Y_grid, matrix, cmap='viridis', edgecolor='k', alpha=0.7)
ax3.plot_wireframe(X, Y_grid, matrix, color='black', linewidth=0.5)
ax3.set_title('Surface + Mesh (surfc) plot')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_zlabel('z')

plt.tight_layout()
plt.show()

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
for i in range(matrix.shape[0]):
    ax.plot(x, [y[i]]*len(x), matrix[i, :], color='b')
ax.set_title('Waterfall plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
````

## Task 5
Create a vector of function values y(x) = \cos(3x) + \cos(x) at points x = -10:0.5:10. Perform interpolation using three available methods. On one plot, show the interpolation nodes and the three resulting function curves. For the given x, y, check the operation of the **polyfit** function using different degrees of the approximating polynomial.

````python
from scipy import interpolate

x_nodes = np.arange(-10, 10.5, 0.5)
y_nodes = np.cos(3 * x_nodes) + np.cos(x_nodes)

x_dense = np.linspace(-10, 10, 1000)

# Interpolation: linear
interp_linear = interpolate.interp1d(x_nodes, y_nodes, kind='linear')
y_linear = interp_linear(x_dense)

# Interpolation: cubic
interp_cubic = interpolate.interp1d(x_nodes, y_nodes, kind='cubic')
y_cubic = interp_cubic(x_dense)

# Interpolation: spline (B-spline)
spl = interpolate.make_interp_spline(x_nodes, y_nodes)
y_spline = spl(x_dense)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x_nodes, y_nodes, 'o', label='Interpolation nodes')
plt.plot(x_dense, y_linear, label='Linear interpolation')
plt.plot(x_dense, y_cubic, label='Cubic interpolation')
plt.plot(x_dense, y_spline, label='B-spline')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Interpolation of function y(x) = cos(3x) + cos(x)')
plt.legend()
plt.grid(True)
plt.show()
````

Perform integration over time $t = 3\,\text{s}$.

Since the equation is first-order, both the state variable vector and the initial condition vector will be single-element (in other words - they will be numbers).

````python
from scipy.integrate import solve_ivp

def dydt(t, y):
    return - (1 + np.sin(y)**2) * y

t_span = (0, 3)
t_eval = np.linspace(0, 3, 300)

initial_conditions = [0.5, 1.0, 2.0, -1.0]
plt.figure(figsize=(8, 5))
for y0 in initial_conditions:
    sol = solve_ivp(dydt, t_span, [y0], t_eval=t_eval)
    plt.plot(sol.t, sol.y[0], label=f'y(0)={y0}')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solutions to the differential equation for different y(0)')
plt.legend()
plt.grid(True)
plt.show()
````
